### Business Problem Understanding

**Context**

Auto insurance company atau perusahaan yang bergerak di bidang asuransi mobil mengalami peningkatan seiring bertumbuhnya industri otomotif dan peningkatan kepemilikan kendaraan.
Mordor Intelligence menyebutkan bahwa pasar ini akan terus tumbuh dengan Compound Annual Growth Rate sekitar 6-7% beberapa tahun ke depan. Selain itu, berdasarkan studi, global motor insurance market akan mengalami peningkatan dari USD 2.13 T menjadi USD 2.95 T pada tahun 2030 (Mordor Intelligence, 2025). Atas fakta tersebut, persaingan antar perusahaan penyedia jasa asuransi kendaraan diprediksi akan semakin ketat. Dalam kondisi ini, memahami perilaku pelanggan menjadi kunci untuk menjaga keberlangsungan suatu bisnis utamanya adalah bisnis asuransi otomotif.

Salah satu indikator penting yang digunakan perusahaan asuransi adalah **Customer Lifetime Value (CLV), sebuah metrik yang digunakan untuk mengestimasi nilai keuntungan yang dapat diperoleh perusahaan dari seorang pelanggan sepanjang hubungan bisnis antara pelanggan dan perusahaan yang terjalin (Gupta & Lehmann, 2003).** Pemahaman dataset **CLV sangat krusial bagi perusahaan asuransi kendaraan utamanya bagi tim marketing, finance, dan customer relationship management** sebelum membuat strategi pemasaran, retensi pelanggan, serta kebijakan alokasi sumber daya.

Melalui CLV, perusahaan asuransi dapat menggunakan metrik tersebut untuk **memahami kebutuhan pelanggannya, menentukan budget marketing, prediksi pendapat jangka panjang, dan membantu pengambilan keputusan sebelum membangun strategi** investasi baik pada pelanggan loyal mereka maupun pelanggan baru di masa yang akan datang.



**Problem Statement**

Salah satu tantangan yang dihadapi oleh perusahaan penyedia auto insurance adalah penentuan strategi bisnis dan retensi pelanggan yang dapat meningkatkan total revenue perusahaan. Strategi bisnis perlu dirancang sedemikian rupa dengan mengacu pada database yang telah dimiliki oleh perusahaan. Namun, tantangannya adalah perusahaan perlu mengestimasi CLV untuk mengambil insight dari data yang dimiliki untuk kebutuhan strategi bisnis. Hal ini cukup krusial karena dapat **menimbulkan risiko berupa kesalahan dalam pengambilan keputusan** bisnis yang dapat membuat perusahaan mengalami kerugian apabila keputusan tidak berdasarkan data dan estimasi yang benar. **Tanpa estimasi CLV yang tepat, perusahaan bisa over-invest** pada pelanggan yang bernilai rendah atau **under-invest pada pelanggan premium**, yang dapat **mengurangi profitabilitas total** hingga puluhan persen dari pendapatan potensial. Permasalahan tersebut juga dapat **menyebabkan penurunan ROI marketing**, **peningkatan churn rate**, dan **kehilangan peluang pertumbuhan profit** dari pelanggan premium. Selain itu, perusahaan mengalami tantangan dalam menentukan pelanggan mana yang memberikan kontribusi profit terbesar selama periode tertentu.

Menjaga dan meningkatkan profitabilitas pelanggan jangka panjang merupakan sebuah tantangan yang besar. Perusahaan perlu mengidentifikasi pelanggan bernilai tinggi dan mencegah pelanggan tersebut berpindah ke kompetitor sekaligus mengelola pelanggan bernilai rendah agar tetap berkontribusi terhadap revenue perusahaan.  

**Goals**

Atas permasalahan tersebut, perusahaan penyedia jasa asuransi kendaraan perlu **membangun model machine learning yang dapat memprediksi CLV** sebagai metrik dasar dalam mengestimasi biaya yang perlu dikeluarkan untuk menjaga loyalitas customer ataupun mendapatkan customer baru serta memprediksi revenue yang akan diperoleh di masa mendatang sehingga perusahaan nantinya mampu menyusun strategi bisnis yang tepat. **Model ini diharapkan dapat menghasilkan galat < 20% sehingga keputusan bisnis berdasarkan hasil prediksi dapat dipercaya.**
Perusahaan asuransi perlu memiliki beberapa fitur penting yang dapat digunakan untuk memprediksi nilai CLV. Fitur-fitur yang dimaksud antara lain informasi demografi (status pekerjaan, status pernikahan, tingkat pendidikan, dan pendapatan), variabel keuangan (monthly premium auto, total claim amount), tipe kendaraan, coverage insurance, dan sebagainya untuk menentukan CLV sebagai target yang akan diprediksi nilainya.

**Analytic Approach**

Untuk mencapai tujuan tersebut, maka perlu untuk melakukan pemilihan fitur serta menganalisis pola dari masing-masing fitur yang berpengaruh terhadap prediksi nilai CLV dengan berdasarkan domain knowledge terkait.

Kemudian, sebuah model regresi dibangun untuk membantu perusahaan dalam memprediksi nilai CLV setiap pelanggan di masa mendatang yang mana akan berguna bagi perusahaan untuk membangun strategi bisnis, menentukan biaya pengeluaran untuk menjaga dan meraih pelanggan baru, serta prediksi revenue yang diperoleh perusahaan. Fitur-fitur yang ada akan dijadikan sebagai data uji bagi model regresi.

**Metric Evaluation**

Evaluasi model regresi untuk prediksi Customer Lifetime Value (CLV) dilakukan dengan beberapa metrik, seperti MAE, medianAE, dan MAPE untuk mengukur tingkat akurasi prediksi. Kemudian, RÂ² digunakan untuk melihat seberapa besar variasi CLV yang dapat dijelaskan model. Masing-masing metriks tersebut akan menghasilkan suatu nilai yang nantinya akan digunakan untuk mengevaluasi masing-masing model regresi yang telah dibangun dan model regresi yang memiliki evaluasi metrik terbaik akan dipilih sebagai tool yang dipercaya perusahaan untuk memprediksi CLV masa mendatang.

#### Perbandingan Mean dan Median per-fitur

Berikut merupakan hasil analisis nilai mean dan median pada fitur kategorikal untuk setiap segmennya yang dilakukan baik secara numerik maupun secara visualisasi grafik.

```python
categorical_features = [
    'Vehicle Class',
    'Coverage',
    'Renew Offer Type',
    'EmploymentStatus',
    'Marital Status',
    'Education'
]

# Loop through each categorical feature
for col in categorical_features:
    stats = df.groupby(col)['Customer Lifetime Value'].agg(['mean', 'median']).sort_values('mean', ascending=False)
    print(f"\nð CLV Mean & Median by {col}:\n")
    display(stats)
```

```python
# Define the grouped mean & median CLV DataFrames
grouped_data = {
    "Vehicle Class": df.groupby("Vehicle Class")["Customer Lifetime Value"].agg(["mean","median"]).reset_index(),
    "Coverage": df.groupby("Coverage")["Customer Lifetime Value"].agg(["mean","median"]).reset_index(),
    "Renew Offer Type": df.groupby("Renew Offer Type")["Customer Lifetime Value"].agg(["mean","median"]).reset_index(),
    "EmploymentStatus": df.groupby("EmploymentStatus")["Customer Lifetime Value"].agg(["mean","median"]).reset_index(),
    "Marital Status": df.groupby("Marital Status")["Customer Lifetime Value"].agg(["mean","median"]).reset_index(),
    "Education": df.groupby("Education")["Customer Lifetime Value"].agg(["mean","median"]).reset_index(),
}

# Plot
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15,15))
axes = axes.flatten()

for ax, (feature, data) in zip(axes, grouped_data.items()):
    data_melted = data.melt(id_vars=feature, value_vars=["mean","median"],
                            var_name="Statistic", value_name="CLV")
    sns.barplot(
        x=feature,
        y="CLV",
        hue="Statistic",
        data=data_melted,
        palette="Set2",
        ax=ax
    )
    ax.set_title(f"CLV Mean vs Median by {feature}")
    ax.set_xlabel(feature)
    ax.set_ylabel("CLV")
    ax.tick_params(axis="x", rotation=45)

plt.tight_layout()
plt.show()
```

Perbandingan nilai rata-rata dan median CLV dilakukan baik secara numerik maupun secara visualisasi. Analisis lebih lanjut dilakukan secara numerik berdasarkan perhitungan dengan metode IQR (Interquartile Range) yang didemonstrasikan pada tahap berikut.

```python
# Deteksi outlier numerik dengan IQR
print("\n===== Deteksi Outlier (berdasarkan IQR) =====\n")
numeric_cols = df_model_cleaned.select_dtypes(include=["float64", "int64"]).columns
for col in numeric_cols:
    Q1 = df_model_cleaned[col].quantile(0.25)
    Q3 = df_model_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = ((df_model_cleaned[col] < lower_bound) | (df_model_cleaned[col] > upper_bound)).sum()
    print(f'{col}')
    print(f"lower bound : {lower_bound}")
    print(f"upper bound : {upper_bound}")
    print(f"Total : {outliers} outlier\n")
```

Hasil perhitungan IQR menunjukkan bahwa hanya fitur income yang tidak memiliki outlier dan ini mengindikasikan bahwa fitur income memiliki distribusi data yang stabil sedangkan sisanya memiliki outlier setidaknya lebih dari 200 data. Transformasi fitur numerik dilakukan untuk fitur MPA dan Income

```python
from sklearn.preprocessing import PowerTransformer, StandardScaler

# --- Transformasi MPA dengan Box-Cox (harus > 0, jika ada nol tambahkan +1) ---
pt_mpa = PowerTransformer(method="box-cox")
X["MPA_boxcox"] = pt_mpa.fit_transform(X[["Monthly Premium Auto"]])

# --- Transformasi Income dengan Yeo-Johnson (bisa nol/negatif) ---
pt_income = PowerTransformer(method="yeo-johnson")
X["Income_yj"] = pt_income.fit_transform(X[["Income"]])

# # --- Transformasi TCA dengan Box-Cox ---
# pt_tca = PowerTransformer(method="box-cox")
# X["TCA_boxcox"] = pt_tca.fit_transform(X[["Total Claim Amount"]])

# --- Standardisasi semua fitur hasil transformasi ---
scaler = StandardScaler()
X[["MPA_scaled", "Income_scaled"]] = scaler.fit_transform(
    X[["MPA_boxcox", "Income_yj"]]
)

# --- Drop kolom aslinya (opsional) ---
X = X.drop(columns=[
    "Monthly Premium Auto", "Income","MPA_boxcox", "Income_yj"
])
```

Fitur numerik Monthly Premium Auo (MPA) dan Income telah melalui proses transformasi untuk mengurangi skewness dan menyesuaikan distribusi data. MPA dilakukan dengan melibatkan box-cox menjadi MPA_boxcox untuk mengurangi skewness yang tinggi, sedangkan Income diberlakukan dengan Yea-Johnson transform dan hasil nya disimpan dalam variabel Income_yj agar distribusi lebih normal dan tetap dapat menangani nilai 0.

Seluruh fitur numerik distandardisasi dengan menggunakan StandardScaler agar memiliki mean mendekati 0 dan standar deviasi mendekati angka 1 (lihat pada tabel dibawah) sehingga siap untuk dimasukkan ke dalam mdel regresi yang sensitif terhadap skala fitur. ### Model Benchmark

Pemodelan prediksi Customer Lifetime Value (CLV) melibatkan beberapa algoritma machine learning, yaitu Linear Regression, K-Nearest Neighbors (KNN), Decision Tree, Random Forest, XGBoost, dan LightGBM. Algoritma tersebut akan diuji performanya menggunakan data latih yang telah disiapkan dan dibandingkan performanya berdasarkan galat yang dihasilkan.

Sebelum dilakukan evaluasi, variabel target (CLV) telah melalui proses transformasi menggunakan Box-Cox Transformation untuk mengurangi skewness dan mendekatkan distribusi data pada normalitas. Namun, agar hasil evaluasi tetap bermakna dalam skala bisnis, nilai target dan prediksi dikembalikan lagi ke skala aslinya melalui inverse transform. Asumsi bahwa CLV dalam mata uang USD.

Proses pelatihan model menggunakan teknik K-Fold Cross Validation dengan 5 lipatan. Pada setiap lipatan, data dibagi menjadi train set dan test set. Model dilatih pada train set, kemudian diuji pada test set untuk menghasilkan prediksi. Hasil prediksi dibandingkan dengan data aktual untuk menghitung metrik evaluasi.

Metrik yang digunakan dalam evaluasi meliputi: MAE, MAPE, MedianAE, R2

Setelah dilakukan evaluasi pada masing-masing fold, hasil metrik dihitung rata-rata dan standar deviasinya. Rata-rata digunakan untuk melihat performa umum model, sedangkan standar deviasi digunakan untuk melihat konsistensi model antar fold.

Seluruh hasil disajikan dalam bentuk tabel dan diurutkan berdasarkan MAE rata-rata (MAE Mean). Pengurutan ini bertujuan agar model dengan kesalahan prediksi paling rendah dapat langsung diidentifikasi sebagai kandidat terbaik.

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, median_absolute_error, mean_absolute_percentage_error
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.preprocessing import PowerTransformer


# Kandidat model
models = {
    "Linear Regression": LinearRegression(),
    "KNN": KNeighborsRegressor(),
    "Decision Tree": DecisionTreeRegressor(random_state=1),
    "Random Forest": RandomForestRegressor(random_state=1),
    "XGBoost": XGBRegressor(random_state=1, verbosity=0),
    "LightGBM": LGBMRegressor(random_state=1)
}

cv = KFold(n_splits=5, shuffle=True, random_state=1)
results = []

for name, model in models.items():
    mae_list, mape_list, r2_list, median_list = [], [], [], []

    for train_idx, test_idx in cv.split(X_train):
        X_tr = X_train.iloc[train_idx]
        X_te = X_train.iloc[test_idx]
        y_tr = y_train.iloc[train_idx]
        y_te = y_train.iloc[test_idx]

        model.fit(X_tr, y_tr)
        y_pred_fold = model.predict(X_te)

        # Inverse transform ke skala asli CLV
        y_te_orig = pt_y.inverse_transform(y_te.values.reshape(-1,1)).flatten()
        y_pred_orig = pt_y.inverse_transform(y_pred_fold.reshape(-1,1)).flatten()

        # Hitung metrik per fold
        mae_list.append(mean_absolute_error(y_te_orig, y_pred_orig))
        mape_list.append(mean_absolute_percentage_error(y_te_orig, y_pred_orig))
        r2_list.append(r2_score(y_te_orig, y_pred_orig))
        median_list.append(median_absolute_error(y_te_orig, y_pred_orig))

    results.append({
        "Model": name,
        "MAE Mean": np.mean(mae_list),
        "MAE Std": np.std(mae_list),
        "MAPE Mean": np.mean(mape_list),
        "MAPE Std": np.std(mape_list),
        "MedianAE Mean": np.mean(median_list),
        "MedianAE Std": np.std(median_list),
        "R2": np.mean(r2_list)
    })

df_results = pd.DataFrame(results)
df_results = df_results.sort_values(by="MAE Mean")
display(df_results)
```

Nilai RÂ² yang diperoleh kedua model mencapai sekitar 0,66, yang berarti mampu menjelaskan 66% variasi data CLV. Hal ini mengindikasikan bahwa kedua algoritma ensemble tree tersebut mampu menangkap hubungan non-linear antar fitur dengan baik serta memberikan prediksi yang relatif konsisten.

Sementara itu, model XGBoost juga menunjukkan performa yang cukup baik dengan MAE sekitar 1663 dan RÂ² sebesar 0,64, meskipun tidak sebaik Random Forest dan LightGBM. Adapun model Decision Tree dan KNN memiliki performa lebih rendah dengan nilai RÂ² masing-masing sebesar 0,36 dan 0,48. RandomForest GridSearch

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import make_scorer, mean_absolute_error, mean_absolute_percentage_error, r2_score, median_absolute_error

# =====================
# 1. Model dasar
# =====================
rf = RandomForestRegressor(random_state=42)

# =====================
# 2. Grid Hyperparameter lebih luas
# =====================
param_grid = {
    "n_estimators": [100, 300, 500, 1000],       # jumlah pohon lebih banyak
    "max_depth": [None, 10, 20, 30, 50],         # variasi kedalaman lebih luas
    "min_samples_split": [2, 5, 10],             # minimal sampel untuk split
    "min_samples_leaf": [1, 2, 4, 8],            # minimal sampel per daun
    "max_features": ["sqrt", "log2"]             # jumlah fitur tiap split
}

# =====================
# 3. Definisikan Scoring
# =====================
scoring = {
    "MAE": make_scorer(mean_absolute_error, greater_is_better=False),
    "MAPE": make_scorer(mean_absolute_percentage_error, greater_is_better=False),
    "R2": make_scorer(r2_score),
    "MedianAE": make_scorer(median_absolute_error, greater_is_better=False)
}

# =====================
# 4. Cross Validation + GridSearchCV
# =====================
cv = KFold(n_splits=5, shuffle=True, random_state=42)

grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=cv,
    scoring=scoring,
    refit="MAE",   # pakai MAE untuk pilih model terbaik
    n_jobs=-1,
    verbose=2
)

# =====================
# 5. Fit ke data
# =====================
grid_search.fit(X_train, y_train)

# =====================
# 6. Hasil terbaik
# =====================
print("Best Parameters:", grid_search.best_params_)
print("Best MAE (CV):", -grid_search.best_score_)

# =====================
# 7. Evaluasi pada test set
# =====================
best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X_test)

# Inverse transform (skala asli CLV)
y_test_orig = pt_y.inverse_transform(y_test.values.reshape(-1,1)).flatten()
y_pred_orig = pt_y.inverse_transform(y_pred.reshape(-1,1)).flatten()

print("MAE:", mean_absolute_error(y_test_orig, y_pred_orig))
print("MAPE:", mean_absolute_percentage_error(y_test_orig, y_pred_orig))
print("MedianAE:", median_absolute_error(y_test_orig, y_pred_orig))
print("RÂ²:", r2_score(y_test_orig, y_pred_orig))
```
